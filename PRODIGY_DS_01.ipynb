```python
import os
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("ads.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.fillna(0)
```

```python
sns.barplot(data=df, x="Gender", y="Age", hue="Purchased")
plt.title("PURCHASE OF MALES V/S FEMALES WITH RESPECT TO AGE")
```

```python
sns.barplot(data=df, x="Age", y="Purchased", hue="Gender")
plt.title("PURCHASE AT EVERY AGE OF MALES & FEMALES")
```

```python
sns.barplot(data=df, y="Age", x="Purchased", hue="Gender")
plt.title("PURCHASE OF MALES & FEMALES AT DIFFERENT AGES")
```

```python
sns.histplot(data=df, x="Gender", y="Age", hue="Purchased")
plt.title("PURCHASE OF MALES V/S FEMALES WITH RESPECT TO AGE")
```

```python
sns.histplot(data=df, x="Age", y="Purchased", hue="Gender")
plt.title("PURCHASE AT EVERY AGE OF MALES & FEMALES")
```

```python
sns.histplot(data=df, y="Age", x="Purchased", hue="Gender")
plt.title("PURCHASE OF MALES & FEMALES AT DIFFERENT AGES")
```
